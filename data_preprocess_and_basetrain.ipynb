# Load Test dataset AgNews

```python
import pandas as pd


Ag_news_test = pd.read_csv("Data_Raw_File/raw/ag_news/ag_news_csv/test.csv",sep = ",")
Ag_news_test.columns = ["Article Class", "Title", "Description"]
Ag_news_test 

# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.map.html

# If labels are already numeric (0,1,2,3), mapping is not required for training, but label2id/id2label is useful for human-readable outputs.
# I originally used maping changes but it gives  error for
#  labels.append(self.Model.label2id[label])
# KeyError: 'Sci/Tech'


mapping_mask = {1:"World" ,2:"Sports",3:"Business",4:"Sci/Tech"}
Ag_news_test["Article Class"] =  Ag_news_test["Article Class"].map(mapping_mask)
Ag_news_test
```

#Basic Test dataset AgNews with all all labels((This is for training and testing how text classfication work))

```python
NLP_base_test = pd.DataFrame({
    "text" : Ag_news_test["Description"],
    "label":  Ag_news_test["Article Class"]
})

NLP_base_test.to_csv("NLP_dataset/Base_test_AgNews.tsv", sep= "\t", index=False,encoding="utf-8")


```
